```python
import urllib.request as req
import bs4
import csv

with open("pttdata.csv", "w", newline="", encoding='UTF-8') as csvfile:
  writer = csv.writer(csvfile)
  writer.writerow(["標題", "內文"])
  
  url = "https://www.ptt.cc/bbs/TOEFL_iBT/index.html"
  #建立一個Request的物件，附加 Requesst Header的資訊
  request = req.Request(url, headers = {
      "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
  })
  with req.urlopen(request) as response:
    data = response.read().decode("utf-8")
  for i in range(1465, 1490, 1):
    root = bs4.BeautifulSoup(data, "html.parser")#讓BeautifulSoup協助我們解析HTML格式文
    links = root.find_all("div", class_="title")#文章標題
    
    links = [x for x in links if "心得" in (" " + x.text + " ")]
            
    
    for link in links:
      url = "https://www.ptt.cc"+link.a["href"]
      request = req.Request(url, headers = {
          "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
      })
      with req.urlopen(request) as response:
        data = response.read().decode("utf-8")
      root = bs4.BeautifulSoup(data, "html.parser")
      ## PTT 上方4個欄位（作者、看版、標題、日期）
      header = root.find_all('span','article-meta-value')
      # 標題
      title = header[2].text
      
      ## 查找所有html 元素 抓出內容
      main_container = root.find(id='main-container')
      # 把所有文字都抓出來
      all_text = main_container.text
      # 把整個內容切割透過 "-- " 切割成2個陣列
      pre_text = all_text.split('--')[0]
      # 把每段文字根據 '\n' 切開
      texts = pre_text.split('\n')
      # 前2個list為空白及標題，直接拿掉
      contents = texts[2:]
      # 內容
      content = '\n'.join(contents)
      # 顯示
      print('標題：'+title)
      print('內容：'+content)
      
      writer.writerow([title, content])

    url = "https://www.ptt.cc/bbs/TOEFL_iBT/index"+ str(i) +".html"
    #建立一個Request的物件，附加 Requesst Header的資訊
    request = req.Request(url, headers = {
        "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/107.0.0.0 Safari/537.36"
    })
    with req.urlopen(request) as response:
        data = response.read().decode("utf-8")
    print("-------------------------------------------------------換頁---------------------------------------------------------")
    print()
```
